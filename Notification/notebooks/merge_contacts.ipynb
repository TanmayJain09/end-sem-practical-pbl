### **1. Importing Libraries**

```python
import pandas as pd
from pathlib import Path
```

### **2. Define Project Paths**

```python
#get project root
PROJECT_ROOT = Path().resolve().parent
```

```python
#data directories
RAW_DIR = PROJECT_ROOT / "data" / "raw"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
```

```python
#file paths
CRITICAL_FILE = PROCESSED_DIR / "critical_escalation_table.csv"
CONTACT_FILE = RAW_DIR / "contact_details.xlsx"
MERGED_FILE = PROCESSED_DIR / "critical_contacts.csv"
```

### **3. Loading Data**

```python
critical_df = pd.read_csv(CRITICAL_FILE)
contacts_df = pd.read_excel(CONTACT_FILE)
```

### **4. Merge Data on PRN and Name**

We would merge on the bases of the *PRN* and the *Name* cause they are the unique and same values in both the data

```python
merged_df = pd.merge(
    critical_df,
    contacts_df,
    on = ["PRN", "Name"],
    how = "left"
)
```

### **5. Cleaning the merged_df**

In this we would drop the unnecessary columns and only keep the required : 
- *risk_category* and *performance_score* is unnecessary
- *PRN, Name, Lectures_Attended, Total_Lectures_Batch* and *Net_Attendance_Percentage* are required for the custom message
- *email, phone* and *whatsapp* are for the contacts

```python
columns_to_drop = ["risk_category","performance_score"]
merged_df = merged_df.drop(
    columns = [
        col for col in columns_to_drop 
            if col in merged_df.columns
    ]
)
```

### **6. Save merged csv**

```python
merged_df.to_csv(MERGED_FILE, index=False)
```
